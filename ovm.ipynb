```python
from ovm import ovm
import matplotlib.pyplot as plt
import numpy as np
```

```python
model = ovm(max_speed=3.0) # creating the model
model.integrate()
```

```python
# plotting model V function
x = np.arange(0.0, 10.0, 0.1)
y = model.V(x)
plt.plot(x, y)
plt.show()
```

```python
jump_car = 1 # The bigger, the less cars we see on the graph

x = model.x
time = model.time
plt.figure(figsize=(20, 10))

start, end = None, None # start and end limits (to zoom)

for j in np.arange(0, model.N, jump_car):
    r = x[j,start:end]
    plt.plot(time[start:end], r, lw=0.3, c="k")
plt.show()
```
